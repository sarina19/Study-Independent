MENYIAPKAN DATASET

(https://github.com/rismamufa/Studi_Independen/tree/konsumen2/Orbit%20Academy)

```python
# Menyiapkan data dan memanggil data set dari github

Dataset = pd.read_csv('https://raw.githubusercontent.com/rismamufa/Studi_Independen/konsumen2/Orbit%20Academy/konsumen2.csv', error_bad_lines=False)
Dataset.keys()
```

```python
# Menampilkan 5 baris data pertama dari dataset tersebut

dataku = pd.DataFrame(Dataset)
dataku.head()
```

```python
# konversi ke data array

X = np.array(Dataset)
print (X)
```

```python
# Menampilkan data dalam bentuk scatter plot

plt.scatter(X[:,0],X[:,1], label='True Position')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik konsumen")
plt.show()
```

**LANGKAH 4**

**MENGGUNAKAN LIBRARY K-MEANS**

**Tulis codingan ini**

image.png

**Tampilan run**

image.png

```python
# Mengaktifkan K Means dengan jumlah K = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

**LANGKAH 5**

**MENAMPILKAN OUTPUT**

```python
# Plot data point
#Memvisualisasikan bagaimana telah di klasterisasikan

plt.scatter(X[:,0],X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik konsumen")
plt.show()
```

```python
# Plot data point
# Memvisualisasikan bagaimana telah di klasterisasikan
# Menampilkan centroids dengan warna hitam
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_,
		cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],
		kmeans.cluster_centers_[:,1],
		color='black')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Konsumen")
plt.show()
```
